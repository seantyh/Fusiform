```python
def run_text_to_avg_char(fpath):
    with open(fpath, "r", encoding="UTF-8") as fin:
        text = fin.read()
    
    digestor = Fusiform.Digestor(Fusiform.spotlight.spotlight_rectangle, 
                                 Fusiform.steps.step_char_wise)
    transX = Fusiform.aggregate_average(text, digestor, 
                    xoffset=0, nchar=1, step_overlap=False)
    plt.imshow(transX, cmap="gray")
```

## Bigram (charac boundary)

```python
def run_text_to_avg_char(fpath):
    with open(fpath, "r", encoding="UTF-8") as fin:
        text = fin.read()
    
    digestor = Fusiform.Digestor(Fusiform.spotlight.spotlight_rectangle, 
                                 Fusiform.steps.step_char_wise)
    transX = Fusiform.aggregate_average(text, digestor, 
                    xoffset=0, nchar=2, step_overlap=False)
    plt.imshow(transX, cmap="gray")
```

## Bigram (inter-charac)

```python
def run_text_to_avg_interchar(fpath):
    with open(fpath, "r", encoding="UTF-8") as fin:
        text = fin.read()
    
    digestor = Fusiform.Digestor(Fusiform.spotlight.spotlight_rectangle, 
                                 Fusiform.steps.step_char_wise)
    transX = Fusiform.aggregate_average(text, digestor, 
                    xoffset=0.5, nchar=2, step_overlap=False)
    plt.imshow(transX, cmap="gray")
```
