```python
txt = "如何讓我遇見你在我最美好的時刻"
txt_w, txt_h = Fusiform.measure_text(txt)
im = Fusiform.text2bitmap(txt, (txt_w, txt_h+1))
plt.imshow(im)
```

```python
def run_text_to_avg(fpath):
    with open(fpath, "r", encoding="UTF-8") as fin:
        text = fin.read()
    transX = Fusiform.aggregate_average(text)
    plt.imshow(transX, cmap="gray")
```

```python
def get_resource(fname):
    return os.path.join("../05_Resource/short_text", fname)
```
